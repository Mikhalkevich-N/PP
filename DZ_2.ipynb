```python
def sum_distance(from_num, to_num):
    # Проверяем, не больше ли первое число второго
    if from_num > to_num:
        # Если да, меняем числа местами
        from_num, to_num = to_num, from_num
    
    # Используем формулу суммы арифметической прогрессии
    # Сумма = (n * (a1 + an)) / 2
    # где n - количество элементов
    # a1 - первый элемент
    # an - последний элемент
    n = to_num - from_num + 1  # количество чисел в диапазоне
    total_sum = n * (from_num + to_num) // 2
    
    return total_sum

# Примеры использования:
print(sum_distance(1, 7))   # Выведет 15 (1+2+3+4+5)
print(sum_distance(7, 1))   # Выведет 15 (автоматически поменяет местами)
print(sum_distance(3, 3))   # Выведет 3 (только одно число)
```

```python
def trim_and_repeat(text: str, offset: int = 0, repetitions: int = 1) -> str:  
    # Обрезаем строку слева на указанное количество символов  
    trimmed_text = text[offset:]  
    # Повторяем строку нужное количество раз  
    result = trimmed_text * repetitions  
    return result  
  
# Примеры использования:  
print(trim_and_repeat("hello"))  # -> "hello"  
print(trim_and_repeat("hello", 2))  # -> "llo"  
print(trim_and_repeat("hello", 1, 3))  # -> "elloeelloeello"  
```
