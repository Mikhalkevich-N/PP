```python
#выполнялось на кали
import subprocess
from collections import Counter

file_path = 'dhcp.pcapng'

# Выполнение анализа DNS-запросов
command_dns_queries = f"tshark -r {file_path} -Y dns -T fields -e dns.qry.name | sort | uniq -c | sort -rn | head -20"
output_dns_queries = subprocess.check_output(command_dns_queries, shell=True, text=True)

# Обработка результатов
lines = output_dns_queries.strip().split('\\n')
counter = Counter()
for line in lines:
    if line.strip():  # Проверка на пустую строку
        count, domain = line.strip().split(maxsplit=1)
        counter[domain] += int(count)

# Печать наиболее часто запрашиваемых доменов
print("\\nНаиболее частые DNS-запросы:")
for domain, count in counter.most_common(20):
    print(f"{count}\\t{domain}")

# Другие полезные команды для дальнейшего анализа
commands = [
    # Анализ DHCP-событий
    f"tshark -r {file_path} -Y bootp -V | grep -E 'Message type|Client IP|Your IP' | head -30",
    # Статистика по протоколам
    f"tshark -r {file_path} -q -z io,phs",
]

# Запуск остальных команд
for cmd in commands:
    output = subprocess.check_output(cmd, shell=True, text=True)
    print(f"\\nРезультат выполнения команды '{cmd}'\\n{output}")

```
