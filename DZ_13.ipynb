```python
import os
import requests
import json

# Переменная среды для хранения API ключа
API_KEY = os.getenv("VT_API_KEY")

# Адрес API VirusTotal
BASE_URL = "https://www.virustotal.com/api/v3/"

def authorize_and_request(url):
    """
    Функция выполняет авторизацию и отправляет GET-запрос к API.
    
    param url: Строка с полным URL запроса
    return: Ответ в формате JSON
    """
    # Устанавливаем заголовки с API-кодом
    headers = {
        "accept": "application/json",
        "x-apikey": API_KEY
    }

    # Отправляем запрос
    response = requests.get(url, headers=headers)

    # Проверяем успешность запроса
    if response.status_code != 200:
        print(f"[ERROR]: Запрос завершился неудачей ({response.status_code})")
        return None

    # Возвращаем JSON-данные
    return response.json()

def main():
    """
    Основная функция: получаем отчёт о статусе сканирования файла
    """
    # Реальный хеш файла file_check
    FILE_HASH = "d8693d00983ba312e3e2f158536f814c71fa5d9a6cccc8014551c6604dd4c18b"
    
    # Формируем конечную точку URL
    endpoint_url = f"{BASE_URL}files/{FILE_HASH}"

    # Получаем данные
    result = authorize_and_request(endpoint_url)

    # Выводим результат в консоль
    if result:
        print(json.dumps(result, indent=4))

if __name__ == "__main__":
    # Проверьте установлен ли API KEY в переменной окружения!
    if not API_KEY:
        raise ValueError("Ошибка: установите переменную окружения VT_API_KEY!")
    
    # Запуск основной функции
    main()


```
