```python
import requests

# Отправляем GET-запрос к ресурсу "/posts"
response = requests.get('https://jsonplaceholder.typicode.com/posts')

if response.status_code == 200:
    # Преобразуем полученный JSON в список объектов
    posts = response.json()
    
    # Выводим заголовки и тела первых 5 постов
    for post in posts[:5]:
        print(f'Заголовок: {post["title"]}')
        print(f'Текст поста: {post["body"]}\\n')
else:
    print("Ошибка при выполнении запроса:", response.status_code)
```

```python
import requests

def get_weather(city_name):
    api_key = 'MY_API_KEY'
    base_url = f'http://api.openweathermap.org/data/2.5/weather?q={city_name}&appid={api_key}&units=metric&lang=ru'

    try:
        response = requests.get(base_url)
        
        if response.status_code == 200:
            data = response.json()
            
            temperature = data['main']['temp']
            weather_description = data['weather'][0]['description']
            
            return f"Текущая температура в городе {city_name}: {temperature:.1f}°C\\nОписание погоды: {weather_description}"
        else:
            return f"Ошибка при получении данных ({response.status_code})"
    except Exception as e:
        return str(e)

# Запрашиваем город у пользователя
city = input("Введите название города: ")
print(get_weather(city))
```
